```python
import relation_datalake_api as rda
import anndata 
```

```python
import scanpy as sc
import pandas as pd
```

```python
def split_pipe(x):
    return x.target.split("|")[0]
```

```python
rda.sc_omics.list_available_sc_anndata()
```

```python
data_path_Rd1 = rda.sc_omics.pull_sc_anndata_file('lara_mouse_fibroblast_reseq_2023', '_Rd1')
adata_Rd1 = anndata.read_h5ad(data_path_Rd1)
```

```python
adata_Rd1.obs
```

```python
adata_Rd1.obs['target'] = adata_Rd1.obs['target'].str.replace('|non-targeting', '')
adata_Rd1.obs['target'] = adata_Rd1.obs['target'].str.replace('non-targeting|', '')
adata_Rd1.obs.target = adata_Rd1.obs.apply(split_pipe, axis=1)
adata_Rd1.var['mt'] = adata_Rd1.var_names.str.startswith('mt-')  # annotate the group of mitochondrial genes as 'mt'
adata_Rd1.obs.rename(columns={'replicate':'batch'}, inplace=True)
adata_Rd1.obs['target'] = adata_Rd1.obs['target'].str.replace('TGFbR', 'Tgfbr1')
adata_Rd1.var.index = adata_Rd1.var.index.str.capitalize()
```

```python
adata_Rd1.obs.condition = adata_Rd1.obs.condition.astype('category').cat.rename_categories({'Activated': 'Tgfb1', 'Quiescent': 'Resting'})
```

```python
c_targets = pd.Categorical( adata_Rd1.obs.target.unique() ) 
c_targets
```

```python
c_targets[10:20]
```

```python
sc.pp.calculate_qc_metrics(adata_Rd1, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
adata_Rd1.write_h5ad('Rd1_Data_post_sgRNA_rescue-reseq.h5ad.gzip', compression='gzip')
```

```python
data_path_Rd3 = rda.sc_omics.pull_sc_anndata_file('lara_mouse_fibroblast_reseq_2023', '_Rd3')
adata_Rd3 = anndata.read_h5ad(data_path_Rd3)
```

```python
adata_Rd3.obs['target'] = adata_Rd3.obs['target'].str.replace('|non-targeting', '')
adata_Rd3.obs['target'] = adata_Rd3.obs['target'].str.replace('non-targeting|', '')
adata_Rd3.obs.target = adata_Rd3.obs.apply(split_pipe, axis=1)
adata_Rd3.var['mt'] = adata_Rd3.var_names.str.startswith('mt-')  # annotate the group of mitochondrial genes as 'mt'
adata_Rd3.obs.rename(columns={'replicate':'batch'}, inplace=True)
adata_Rd3.obs['target'] = adata_Rd3.obs['target'].str.replace('TGFbR', 'Tgfbr1')

adata_Rd3.obs['target'] = adata_Rd3.obs['target'].str.replace('EGR2', 'Egr2')

adata_Rd3.var.index = adata_Rd3.var.index.str.capitalize()
```

```python
c_targets = pd.Categorical( adata_Rd3.obs.target.unique() ) 
c_targets
```

```python
sc.pp.calculate_qc_metrics(adata_Rd3, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
adata_Rd3.write_h5ad('Rd3_Data_post_sgRNA_rescue-reseq.h5ad.gzip', compression='gzip')
```
